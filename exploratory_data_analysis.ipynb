# Exploratory Data Analysis with Graduate Admissions Dataset

## Context
This dataset is created for prediction of Graduate Admissions from an Indian perspective.

## Content
The dataset contains several parameters which are considered important during the application for Masters Programs.
The parameters included are :
- GRE Scores ( out of 340 )
- TOEFL Scores ( out of 120 )
- University Rating ( out of 5 )
- Statement of Purpose and Letter of Recommendation Strength ( out of 5 )
- Undergraduate GPA ( out of 10 )
- Research Experience ( either 0 or 1 )
- Chance of Admit ( ranging from 0 to 1 )
- Acknowledgements
- This dataset is inspired by the UCLA Graduate Dataset. The test scores and GPA are in the older format.
- The dataset is owned by Mohan S Acharya.

## Inspiration
This dataset was built with the purpose of helping students in shortlisting universities with their profiles. The predicted output gives them a fair idea about their chances for a particular university.

### Visualization Data Train and Accuracy

```python
range_train_size = np.arange(0, 0.196, 0.001)
train_size  = 0.8
trains_and_accurations = []

for i in range_train_size:
    X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=train_size + i, random_state=10)
    model_linear_regression = LinearRegression()
    model_linear_regression.fit(X_train, y_train)
    y_pred = model_linear_regression.predict(X_test)
    r_squared = float("{:.2f}".format(r2_score(y_test, y_pred)))
    trains_and_accurations.append([round((train_size+i) * 100, 2), round(r_squared, 2) * 100])
```

```python
len(range_train_size)
```

```python
train_and_accuracy_df = pd.DataFrame(data=trains_and_accurations)
```

```python
train_and_accuracy_df = train_and_accuracy_df.rename(columns={0:'Train', 1:'Accuracy'})
```
